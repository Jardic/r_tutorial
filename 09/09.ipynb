```r
test_plot <- function() {
    plot(
        1:10
        , 10 / (1:10)
        , xlab = "x"
        , ylab = "y"
        , xlim = c(0, 10)
        , ylim = c(0, 10)
    )
}

par(bty = "n")
test_plot()
```

```r
par(cex = 1.5)
test_plot()
```

```r
my_params <- list(
    bty = "n"
    , cex = 1.5
    , pch = 15
)
par(my_params)
test_plot()
```

```r
my_params <- c(my_params, list(lty = 2))
par(my_params)
test_plot()
x <- seq(0, 6*pi, pi / 8)
y <- cos(x) + cos(x / 3) + 5
lines(x, y)
```

```r
multi_params <- c(my_params, list(mfrow = c(2, 3)))
par(multi_params)
for (i in 1:6) {
    test_plot()
    rnd_1 <- runif(1, 0, 1)
    rnd_2 <- runif(1, 2, 6)
    x <- seq(0, 6*pi, pi / 8)
    y <- sin(x) + sin(x / rnd_1) + rnd_2
    lines(x, y)
}
```
